**Adapt the code above to generate a 1000 character long string with weights based on the previous two characters.**

As suggested I copied and pasted the below code from Ians information notebook and edited it to generate a 1000 long character string.

```python
# We firstly have to import the relevant module
import urllib.request

# Using the URL of a textbook version of Alice in Wonderland
book_url = 'https://www.gutenberg.org/files/11/11-0.txt'

# Get the book
book = list(urllib.request.urlopen(book_url))

# Decode the lines and strip line endings
book = [line.decode('utf-8-sig').strip() for line in book]

# I then pick a sample paragraph form the book, line 201 to 206 in chapter 1
paragraph = ' '.join(book[201:206])

# printing the sample paragraph
print(paragraph)
```

```python
# Now to manipulate the text in the paragraph

# Change all letters to lowercase
alice = paragraph.lower()

# Include all letters and a space
chars = 'abcdefghijklmnopqrstuvwxyz '

# And strip anything that is not a letter or space
alice = ''.join([c for c in alice if c in chars])

# Print the new paragraph
print(alice)
```

```python
# For doing (pseudo-)random things in Python.
import random

# Print a random character, for example.
print(random.choice(chars))
```

```python
# Get the length of alice.
N = len(alice)

# Generate N random characters from chars.
gener = random.choices(chars, k=N)

# Join them together in a string.
gener = ''.join(gener)

# Print.
print(gener)
```

```python
# Get the whole book in one big string.
sbook = ''.join(book[26:]).lower()

# Create the weights - count the occurences of each character in the whole book.
weights = [sbook.count(c) for c in chars]

# Show the weights.
weights
```

```python
# Generate a string using those weights.
wgenr = random.choices(chars, weights=weights, k=N)

# Join them together in a string.
wgenr = ''.join(wgenr)

# Print.
print(wgenr)
```

```python
# Create the weights.
twoghts = {c: {d: sbook.count(c + d) for d in chars} for c in chars}

# Show the weights.
twoghts
```

```python
# Loop through our character set.
for i in range(len(chars)):
    # Print the character and how many times it appears in Alice in Wonderland.
    print(f'{chars[i]}: {weights[i]}')
```

```python
# Start with space.
pairs = ' '

# creating a 1000 character string
for i in range(1, 1000):
    # Get the weights where the previous character is the last character in twos.
    wt = twoghts[pairs[-1]]
    # Turn wt into a list, ordered by chars.
    wt = [wt[c] for c in chars]
    # Randomly pick the next character using the previous two characters
    nextc = random.choices(chars, weights=wt, k=2)[0]
    # Append the character to twos.
    pairs = pairs + nextc
```

```python
pairs
```

```python
# To check how many characters are printed
len(pairs)
```

## Exercise 2

**Explain why the log of zero is undefined.**

<br/>


Zero doesn’t have a multiplicative inverse as multiplicative inverse is the reciprocal for a number “a”, denoted by 1/a, is a number which when multiplied by “a” yields the multiplicative identity 1.

The multiplicative inverse of a fraction: a/b is b/a and here zero does not have a reciprocal because no real number multiplied by 0 produces 1.

The product of any real number with zero is zero.

So we can say that multiplicative inverse of 0 does not exist or undefined since division by zero is not defined.
